```python
import sqlalchemy as db
import pandas as pd
```

```python
engine = db.create_engine('mysql+pymysql://foxience:lab1.foxience.org@lab1.foxience.org:3306/demo2')
```

```python
conn = engine.connect()
```

```python
meta = db.MetaData()
```

```python
tbl = db.Table('drug_interactions', meta, autoload=True, autoload_with=engine)
```

```python
print(tbl.columns)
pd.set_option('display.max_colwidth', -1)
```

```python
# AcenoCoumarol
# SELECT * FROM drug_interactions WHERE drug1 LIKE '%AcenoCoumarol%'
query = db.select([tbl]).where(tbl.columns.drug1.like('%metformin%'))
print(query)
```

```python
res = conn.execute(query)
pd.DataFrame(res, columns=tbl.columns)
```

```python
# SELECT * FROM drug_interactions WHERE drug1 LIKE '%abc%' AND severe = 1
query = db.select([tbl]).where(tbl.columns.drug1.like('%acetylsalicylic%') & tbl.columns.severe == 1)
res = conn.execute(query)
pd.DataFrame(res, columns=tbl.columns)
```

```python
medications = [
    'warfarin',
    'metformin',
    'prednisolon',
    'clopheniramin',
]
query = db.select([tbl]) \
    .where(tbl.columns.drug1.in_(medications)) \
    .where(tbl.columns.drug2.in_(medications))
res = conn.execute(query)
pd.DataFrame(res, columns=tbl.columns)
```
